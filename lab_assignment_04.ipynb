```python
import pandas as pd
import numpy as np
import statistics
import random


df = pd.read_csv("https://raw.githubusercontent.com/suthapalliuday/INFO5502-Summer2022/main/datasets/lab-assignment-04-student-performance.csv")
df.head()
```

```python
# 1) Description for Columns

for i in df.columns:
  print('Column : ',i)
  print('No. of Records : ',len(df.index))
  print('No. of Unique values for each Feature : ',df[i].value_counts().shape[0])
  print('Frequency For the Unique Values :\n',df[i].value_counts(),'\n')
```

```python
# 2) Visualize the counts of the feature - 'Study Hour'
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(rc = {'figure.figsize':(10,5)})
fig = sns.countplot(x ='sh', data = df,color="salmon", facecolor=(0, 0, 0, 0),
                   linewidth=4,
                   edgecolor=sns.color_palette("Pastel1", 3),)
fig.set(xlabel='Study Hours', ylabel='Count')
for i in fig.patches:    
     fig.text(i.get_x()+i.get_width()/2, i.get_height()+0.75, i.get_height(),  
     horizontalalignment='center',size=9)

plt.show()

#This shows that there are only 20% students who are studying good hours of time i.e., more than 6 hours.
```

```python
# 3) Visualizing the Study hours and the end Semester Percentage (relation between numerical and categorical features)

#convert the categorical variable into numerical 

mod_df = pd.DataFrame(df.sh)
mod_df['numerical_esp'] = df['esp']

for i in df.index:
  if mod_df['numerical_esp'][i] == 'Best':
    mod_df['numerical_esp'][i] = random.randrange(80,101)
  if mod_df['numerical_esp'][i] == 'Good':
    mod_df['numerical_esp'][i] = random.randrange(60,80)
  if mod_df['numerical_esp'][i] == 'Vg':
    mod_df['numerical_esp'][i] = random.randrange(45,60)
  if mod_df['numerical_esp'][i] == 'Pass':
    mod_df['numerical_esp'][i] = random.randrange(30,45)
  if mod_df['numerical_esp'][i] == 'Fail':
    mod_df['numerical_esp'][i] = random.randrange(0,30)

sns.factorplot(x = 'sh', y = 'numerical_esp', kind = 'strip', data = mod_df)

#sns.factorplot(x = 'sh', y = 'esp', data = df)
```

```python
# 4)Correlation and Heatmap
#For getting a correaltion matrix we need to have numerical catagories

from sklearn import preprocessing

def Encoder(df):
    columnsToEncode = list(df.select_dtypes(include=['category','object']))
    enc = preprocessing.LabelEncoder()
    for i in columnsToEncode:
        try:
            df[i] = enc.fit_transform(df[i])
        except:
            print('Problem while encoding '+i)
    return df
var = Encoder(df)
correlation_matrix = var.corr()
sns.set(rc = {'figure.figsize':(18,10)})
sns.heatmap(correlation_matrix, annot = True, cmap = "Blues")
plt.show()
```

```python
#since ms is having almost same values we can just drop it

var = var.drop('ms', axis = 1)
correlation_matrix = var.corr()
sns.set(rc = {'figure.figsize':(18,10)})
sns.heatmap(correlation_matrix, annot = True, cmap = "Blues")
plt.show()
```
